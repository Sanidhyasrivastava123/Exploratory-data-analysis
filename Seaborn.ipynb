```python
'''Seaborn Library:
    1.Used for data visualization and based on matplotlib.
    2.It allows the creation of Statistical Graph
Functionalities:
    1.Allows comparison between multiple algorithms.
    2.Support multi-plot grids.
    3.Available univarate and bivarate visualizations.
    4.Availability of differentcolor palettes.
    5.Estimates and plots Linear Regression automatically.

Matplotlib vs Seaborn
Seaborn fixes two shortcomings of Matplotlib:
    1.Matplotlib settings are diffiult to figure out.It comes with numerous customized themes and high level interfaces.
    2.It does not serve well when it comes to serving Dtaframes.
Python Seaborn Functions:
    1.Visualizing statistical data.
    2.Plotting with categorical data.
    3.Visualisation of distribution of the dataset.'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df=pd.read_csv("C:/Users/new pc/Desktop/Data Sets/Iris.csv")
df.head()
```

```python
# plotting strip plot with seaborn
sns.stripplot(x="SepalLengthCm",y="SepalWidthCm",data=df)
plt.title("Strip Plot")
plt.show()
```

```python
sns.set(style ="whitegrid") 
#.set(style=whitegrid) function here is also use to define the background of plot.We can use darkgrid
#instead of whitegrid if we want dark colored background.

sns.stripplot(x = 'Species', y = 'SepalLengthCm', data = df)
```

```python
#Swarmplot:This is very much similar to striplot but the only difference is that is do not allow overlapping of markers.It cause jittering in the markers of the plot so that graph can easily be readed without information loss as seen in the above plot.
sns.swarmplot(x = 'Species', y = 'SepalLengthCm', data = df)
plt.title('Graph')
```

```python

#KDE stands for Kernel Density Estimation and that is another kind of the plot in seaborn.
#bins is used to set the number of bins you want in your plot and it actually depends on your dataset.
#color is used to specify the color of the plot
sns.distplot(df['SepalLengthCm'],kde = False, color ='red', bins = 30) 
```

```python
sns.jointplot(x="SepalLengthCm",y="SepalWidthCm",data=df) 
```

```python
sns.jointplot(x="SepalLengthCm",y="SepalWidthCm",data=df,kind ='reg') 
```

```python
sns.jointplot(x="SepalLengthCm",y="SepalWidthCm",data=df,kind ='hex') 
```

```python
sns.jointplot(x="SepalLengthCm",y="SepalWidthCm",data=df,kind ='kde') 
```

```python
sns.pairplot(df, hue ="SepalWidthCm", palette ='coolwarm')
```
